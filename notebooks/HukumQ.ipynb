###**<font color="#3685cc"> 2. Prepare Dataset</font>**

```python
import json
import pandas as pd

# Read .json file, and initialize it into a "data" variable
with open("content.json") as data_file:
    data = json.load(data_file)
```

```python
# inserting tags and patterns(input) in data.json into the data frame
text_input = []
tag = []

for intent in data['intents']:
    for pattern in intent['patterns']:
        text_input.append(pattern)
        tag.append(intent['tag'])

df = pd.DataFrame({'text_input': text_input,
                    'tag': tag})
```

##**<font color="#3685cc"> B. Pre-Processing</font>**

###**<font color="#3685cc"> 1. Cleansing</font>**

###**<font color="#3685cc"> 2. Encoding</font>**

```python
df.tag
```

```python
# label encoding
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical

le = LabelEncoder()
y_train = le.fit_transform(df.tag)
print('y_train')
y_train = to_categorical(y_train)
```

```python
y_train
```

###**<font color="#3685cc"> 3. Data Information </font>**

```python
all_vocab = []
length = []

for idx, row in df.iterrows():
    sent = row['text_input']
    [all_vocab.append(i) for i in sent.split()]
    length.append(len(sent.split()))

print("Jumlah semua kata/Vocabulary = ", len(all_vocab))
print("Maksimal jumlah kata dalam satu kalimat = ", max(length))
print("Jumlah kata unik = ", len(set(all_vocab)))
```

##**<font color="#3685cc"> C. Training</font>**

###**<font color="#3685cc"> 1. Modeling</font>**

###**<font color="#3685cc"> a. Prepare Vactorization function </font>**

```python
from tensorflow.keras.layers import TextVectorization

max_vocab_length = len(set(all_vocab))
max_length = max(length)

text_vectorization = TextVectorization(max_tokens=max_vocab_length,
                                       split='whitespace',
                                       ngrams=None,
                                       output_mode='int',
                                       output_sequence_length=max_length
                                       )

text_vectorization.adapt(df.text_input)
```

####**<font color="#3685cc"> b. Emmbedding Layers</font>**

```python
from tensorflow.keras.layers import Embedding
embedding = Embedding(input_dim=max_vocab_length,
                      output_dim=16,
                      embeddings_initializer="uniform",
                      input_length=max_length)
```

####**<font color="#3685cc"> c. Others layers</font>**

```python
from tensorflow.keras import Model
from tensorflow.keras.layers import Input, Dense, LSTM, Bidirectional
inputs = Input(shape=(1,), dtype='string')
x = text_vectorization(inputs)
x = embedding(x)
x = Bidirectional(LSTM(64, return_sequences=True))(x)
x = Bidirectional(LSTM(64))(x)
outputs = Dense(59, activation='softmax')(x)
model_lstm = Model(inputs, outputs, name="LSTM_model")
```

###**<font color="#3685cc"> 2. Compile & Training</font>**

####**<font color="#3685cc"> a. Compile</font>**

```python
import tensorflow as tf

model_lstm.compile(loss='categorical_crossentropy',
                    optimizer=tf.keras.optimizers.legacy.RMSprop(learning_rate=0.0001),
                   metrics=["accuracy"])
```

####**<font color="#3685cc"> b. Training</font>**

```python
train = model_lstm.fit(df.text_input,
               y_train,
               epochs=200,
               verbose=1)
```

```python
model_lstm.evaluate(df.text_input, y_train)
```

```python
model_lstm.save("bot_model.tf")
```

###**<font color="#3685cc"> 3. Model Analysis</font>**

```python
# import library
import json
from string import punctuation
import random
import pickle
from tensorflow.keras.models import load_model

with open("content.json") as data_file:
    data = json.load(data_file)
model = load_model('bot_model.tf')
le_filename = open("label_encoder.pickle", "rb")
le = pickle.load(le_filename)
le_filename.close()

def preprocess_string(string):
    string = string.lower()
    exclude = set(punctuation)
    string = ''.join(ch for ch in string if ch not in exclude)
    return string

def chat(model):
    print("Anda akan dihubungkan ke bot kami, mohon ditunggu")
    exit = False
    while not exit:
        inp = input("Anda : ")
        inp = preprocess_string(inp)
        prob = model.predict([inp])
        results = le.classes_[prob.argmax()]
        if prob.max() < 0.2:
            print("HukumQ : Maaf kak, aku ga ngerti")
        else:
            for tg in data['intents']:
                if tg['tag'] == results:
                    responses = tg['responses']
            if results == '':
                exit = True
                print("END CHAT")
            print(f"HukumQ : {random.choice(responses)}")

if __name__ == "__main__":
    chat(model)
```
